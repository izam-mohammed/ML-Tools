tqdm is a Python library that provides a fast, extensible progress bar for loops and other iterable computations.

```python
# getting the library

from tqdm import tqdm
import time
```

```python
# basic usage

data = range(10)
for item in tqdm(data, desc="Progress items"):
  time.sleep(0.2)
```

```python
# more customization

for item in tqdm(
    data,
    desc="Processing items",
    bar_format="{l_bar}{bar}|  [{elapsed}<{remaining}, {n_fmt}/{total_fmt}, {rate_fmt}]"
    ):

    time.sleep(0.1)
```
