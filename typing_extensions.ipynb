The typing_extensions module in Python provides additional functionality on top of the typing module

```python
# annotated
from typing_extensions import Annotated

value: Annotated[int, "this is a integer"] = 12
value
```

```python
# constant
from typing_extensions import Final

CONST: Final = 43
```

```python
# type
from typing_extensions import Type

class Animal:
    pass

animal_type: Type[Animal] = Animal()
```
