The `typing` module in Python is used to provide type hints and type checking at the time of code development.

but why this module ????

- Improved Readability
- Static Type Checking --> tools like mypy
- IDE Support
- Code Quality and Maintainability
- Enhanced Collaboration

```python
# tuple
from typing import Tuple

def hola() -> Tuple[str, int]:
  name = "izam"
  age = 22
  return name, age

hola()
```

```python
# list
from typing import List

def hola() -> List[int]:
  ls = [2,3,4,5,6]
  return ls

hola()
```

```python
# dict
from typing import Dict

def hola() -> Dict[int, str]:
  data = {1:"hola", 3:"haii"}
  return data

hola()
```

```python
# union --> can have multiple types
from typing import Union

def hola(number: Union[int, float]) -> float:
  return number ** 0.5

hola(9)
```

```python
# optional --> can be of a certain type or None

from typing import Optional

def hola() -> Optional[str]:
  return "hai"

hola()
```

```python
# callable type
from typing import Callable

def perform_operation(operation: Callable[[int, int], int], a: int, b: int) -> int:
    """
    Perform a binary operation using the provided callable.

    Parameters:
    - operation (Callable[[int, int], int]): A callable that takes two integers and returns an integer.
    - a (int): The first operand.
    - b (int): The second operand.

    Returns:
    - int: The result of the operation.
    """
    return operation(a, b)


def multiply(x, y):
    return x * y

perform_operation(multiply, 3, 4)
```

```python
# any -> can have any possible value, including being of any type.

from typing import Any

def mysterious_function(value: Any) -> Any:
    return value

mysterious_function(3)
```
