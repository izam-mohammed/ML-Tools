```python
# getting the heros

import requests
from bs4 import BeautifulSoup
```

```python
# scrapping

url = "https://www.python.org/"
response = requests.get(url)
print(response.text[:1000])
```

it's the entire web page

```python
# understanding the response a bit more
from pprint import pprint

print(response.status_code)
pprint(dict(response.headers))
```

**Manipulating the HTML with BeutifulSoup**

```python
# creating the beutifulsoup obj

soup = BeautifulSoup(response.text, "html.parser")
```

```python
# getting a tag

soup.title
```

```python
# get all links

for i, link in enumerate(soup.find_all('a')):
   print(link.get('href'))
   if i>5:
    print("\nmore ...")
    break
```

```python
# accessing a tag

tag = soup.html
tag.h2
```

```python
# get all p tags

all_p = soup.find_all("p")
len(all_p)
```
