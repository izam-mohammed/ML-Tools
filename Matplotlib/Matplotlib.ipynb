### Getting the libraries

```python
# importing

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Plots

```python
# simple plotting

X = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(X, y)
plt.show()
```

```python
# scatter plot

plt.scatter(X, y)
plt.show()
```

```python
# bar chart

categories = ['A', 'B', 'C', 'D']
values = [10, 24, 17, 30]

plt.bar(categories, values, color='blue')
```

```python
# horizontal barchart

plt.barh(categories, values, color='Red')
```

```python
# histogram

data = np.random.randn(1000)  # generate data

plt.hist(data, bins=20, rwidth=0.8)
plt.show()
```

```python
# pie chart

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
# box plot

data = np.random.randn(1000)

plt.boxplot(data)
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```

```python
# color map

data = np.random.rand(10, 10) #random 2x2 matrix
plt.imshow(data)
plt.colorbar()
```

```python
# error bar

x = [1 , 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]
y_err = [0.5, 0.3, 0.4, 0.2, 0.6]

plt.errorbar(x, y, yerr=y_err, fmt='o-')
plt.show()
```

### Alignment

```python
# alighnment

fig, axes = plt.subplots(2,3, figsize=(8, 5))
plt.tight_layout()
```

```python
# with subplot2grid

fig = plt.figure(figsize=(8, 5)) # initialize figure

ax = [None for _ in range(6)]

ax[0] = plt.subplot2grid((3, 4), (0, 0), colspan=4)
ax[1] = plt.subplot2grid((3, 4), (1, 0), colspan=1)
ax[2] = plt.subplot2grid((3, 4), (1, 1), colspan=1)
ax[3] = plt.subplot2grid((3, 4), (1, 2), colspan=1)
ax[4] = plt.subplot2grid((3, 4), (1, 3), colspan=1, rowspan=2)
ax[5] = plt.subplot2grid((3, 4), (2, 0), colspan=3)

for i in range(6):
    ax[i].set_xticks([])
    ax[i].set_yticks([])

plt.tight_layout()
plt.show()
```

### Customizing the plots

```python
# customise plots

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, label='Line Plot', color='blue', linestyle='--', marker='o')

#labels
plt.text(0.5, 0.5, 'Random text', fontsize=10, color='green')
plt.xlabel('X Label')
plt.ylabel('y Label')
plt.title('Costom plot')

plt.legend() #show the name of each data point
plt.grid(True) #enable grid
```

```python
# adding some limits

x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

plt.plot(x, y)

plt.xlim(0, 10) #x axis limit
plt.ylim(0, 20) #y axis limit
```

```python
# set ticks

x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

plt.plot(x, y)

x_ticks = [1, 2, 3, 4, 5]
x_tick_labels = ['A', 'B', 'C', 'D', 'E']
y_ticks = [2, 4, 6, 8, 10]

plt.xticks(x_ticks, x_tick_labels)  # X-axis custom ticks and labels
plt.yticks(y_ticks)  # Y-axis custom ticks

plt.show()
```

```python
# grid and spine

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y)

plt.grid(True, linestyle='--', alpha=0.5)
ax = plt.gca()
ax.spines['top'].set_visible(False) # dont show top border
ax.spines['right'].set_visible(False)

plt.show()
```

```python
# annotate

x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

plt.plot(x, y)

plt.text(3,6, 'Here it is', fontsize=12, color='red')
```

```python
# arrow annotate

x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

plt.plot(x, y, color='red')

plt.annotate('Here it is', xy=(3,6), xytext=(4, 8),
             arrowprops=dict(facecolor='blue', shrink=0.05))

plt.show()
```
