pytest is a testing framework that simplifies the process of writing and executing tests in Python

**All of the code is available in this [github repo]()**

**What is pytest**

- Testing framework for python
- auto discovery of tests
- rich assertion introspection
- support parameterized and fixture-based testing

**Why Pytest?**

- simplified systax
- rich assrtion introspection
- powerful fixture system
- compatibility
- extensibility
