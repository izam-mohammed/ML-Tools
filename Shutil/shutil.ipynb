```python
import shutil
```

```python
# create a file

def create():
  with open("file.txt", "w"):
    print("created the file")
create()
```

```python
# Copy a file

src_file = 'file.txt'
dest_file = 'sample_data'
shutil.copy(src_file, dest_file)
```

```python
# Copy a directory

src_dir = 'sample_data'
dest_dir = 'data'
shutil.copytree(src_dir, dest_dir)
```

```python
# Remove a directory

dir_to_remove = 'data'
shutil.rmtree(dir_to_remove)
```

```python
# Move a file

src_file = 'file.txt'
dest_file = 'sample_data/hola.txt'
shutil.move(src_file, dest_file)
```

```python
# Move a directory
src_dir = 'sample_data'
dest_dir = 'data'
shutil.move(src_dir, dest_dir)
```

```python
# Get disk usage statistics

path = 'data'
disk_usage = shutil.disk_usage(path)
disk_usage
```

```python
# Create an archive

base_name = 'archive'
format = 'zip'
root_dir = '.'
base_dir = 'data'
shutil.make_archive(base_name, format, root_dir, base_dir)
```

```python
# Unpack an archive

filename = 'archive.zip'
extract_dir = 'hola'
shutil.unpack_archive(filename, extract_dir)
```
