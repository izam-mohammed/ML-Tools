### Getting the libraries

```python
import seaborn as sns

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Getting the data

```python
# importing data set from seaborn itself

crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

### High Level View

![](https://seaborn.pydata.org/_images/function_overview_8_0.png)

### Plots

```python
# distribution plot

sns.displot(crash_df['not_distracted'],kde=True,  bins=25)
plt.plot()
```

```python
# joint plot (for multi variate)

sns.jointplot(x='speeding', y = 'alcohol', data=crash_df, kind='reg')
```

```python
# kde plot

sns.displot(crash_df['alcohol'], kind="kde")
```

```python
# relplot

sns.relplot(data=crash_df, x='speeding', y = 'alcohol', hue="ins_premium")
```

```python
# paiplot (multi variate)

sns.pairplot(crash_df)
```

```python
# changing the data

tips_df = sns.load_dataset('tips')
```

```python
# categorial data in pair plot with hue

sns.pairplot(tips_df, hue='sex', palette='Blues')
```

```python
# bar plot

sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.mean)
```

```python
# count plot

sns.countplot(x = 'sex', data=tips_df)
```

```python
# box plot

sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
# violine plot

sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

```python
# heatmap

crash_mx = crash_df.corr(numeric_only=True)

sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
# changing the data

iris = sns.load_dataset('iris')
```

```python
# pair grid

iris_g = sns.PairGrid(iris, hue='species')
iris_g.map_diag(plt.hist) # using histogram in diagonal
iris_g.map_upper(plt.scatter)  # use scatter plot in upper
iris_g.map_lower(sns.kdeplot)   # use kde plot in lower

plt.legend()
```

```python
# facet grid

tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.hist, 'total_bill', bins=8)
```

```python
# linear plot

sns.lmplot(x='total_bill', y='tip', col='sex', row='time', data=tips_df)
# note that in here, time and sex have only 2 categoris. the datapoints are total bill and tip
```
