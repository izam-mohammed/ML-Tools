## Operations on arrays

```python
# initiating two 1d array

arr1 = np.array([1,2,3])
arr2 = np.array([6,7,8])
```

```python
# some arithmatic ones

print(f"Addition        --> {arr1 + arr2}")
print(f"Substraction    --> {arr1 - arr2}")
print(f"Division        --> {arr1 / arr2}")
print(f"Multiplication  --> {arr1 * arr2}")
print(f"Power           --> {arr1 ** arr2}")
```

```python
# some comparisons

print(f"arr1 values and arr2 values are same      --> {arr1==arr2}")
print(f"arr1 values and arr2 values are not same  --> {arr1!=arr2}")
print(f"arr1 values are greater than arr2 values  --> {arr1>arr2} ")
print(f"arr1 values are less than arr2 values     --> {arr1<arr2}")
```

```python
# some mathematical swimming in arr

print(f"square root of arr1   --> {np.sqrt(arr1)}")
print(f"exponential of arr1   --> {np.exp(arr1)}")
print(f"natural logarithmmmm  --> {np.log(arr1)}")
```

```python
# and some trigenomic things also

print(f"sin of arr1   --> {np.sin(arr1)}")
print(f"cos of arr1   --> {np.cos(arr1)}")
print(f"tan of arr1   --> {np.tan(arr1)}")
# cosec, sec ......
```

```python
# some statistical operations also there

print(f"std   --> {np.std(arr1)}")
print(f"variance   --> {np.var(arr1)}")
```

```python
# linear algebra things

print(f"dot product  --> {arr1 @ arr2}")
print(f"matrix multiplication --> {np.matmul(arr1, arr2)}")
```

## Sorting the Arr

```python
# sorting a 1d array

arr = np.array([7, 12, 4, 6])
arr.sort()
arr
```

```python
# sorting a 2d array based on the row

arr = np.array([[1,34,0],
                [6,2,7]])
arr.sort(axis=1)
arr
```

```python
# sorting a 2d array based on columns

arr.sort(axis=0)
arr
```

```python
# well, what about the argsort, in 1d bytheway

arr = np.array([4,2,7,9,34,74,32,97,1])
arr_indices = arr.argsort()
arr_indices
```

```python
# the use of argsort

arr[arr_indices]
```
